```python
import pandas as pd
india_wheather=pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
india_wheather
```

```python
us_wheather=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
    "humidity":[6,65,75]  
})
us_wheather
```

```python
df=pd.concat([india_wheather,us_wheather])
df
```

```python
df=pd.concat([india_wheather,us_wheather],ignore_index=True) #if we put ignore index then it ignore the index of previos data 
df
```

```python
df=pd.concat([india_wheather,us_wheather],keys=["india","us"]) # keys function associate keys corresponding to each data frame it helps in to find out differnet concat data from data frame  
df
```

```python
df.loc["us"]
```

```python
temperature=pd.DataFrame({
    "city":["delhi","banglore","mumbai"],
    "temperature":[30,25,29]
})
temperature
```

```python
windspeed=pd.DataFrame({
    "city":["delhi","banglore","mumbai"],
    "windspeed":[8,5,7]
})
windspeed
```

```python
#if we use default method to concat two data then it append rows wise but these two data have no meaning if we append in rows
```

```python
df=pd.concat([temperature,windspeed])  # No meaning if we append them as rows
df
```

```python
df=pd.concat([temperature,windspeed],axis=1)
df
```
