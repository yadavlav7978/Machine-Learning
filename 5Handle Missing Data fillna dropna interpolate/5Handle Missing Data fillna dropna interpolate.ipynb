```python
df=pd.read_csv("wheather_data",parse_dates=["day"])  # Here parse_dates makes day string into actual date format 
df
```

```python
df.set_index('day',inplace=True)  # To set day as index
df
```

```python
df1=df.interpolate()  # By using interpulate it some how come with real guess
df1                   # by default it use linear interpulation to guess the value it come with mid value of previous and next value  
```

```python
df1=df.interpolate(method="time") #if we want to guess value by dates then use"time", if we obseve carefully for temperture date
df1                             # 2017-01-05 is more clousre to 4 so guess should come clousre to 28 such as 29 not 30
```
